```python
from pyspark.sql.types import StructType,StructField, StringType, IntegerType,DateType,TimestampType, FloatType
import pyspark.sql.functions as f
catalog_name= "ecommerce"
```

**Load Brand Dimension Table**

```python
#Define schema for data files
brand_schema= StructType([
    StructField("brand_code", StringType(), False),
    StructField("brand_name", StringType(), True),
    StructField("category_code", StringType(), True)
])
#Read data files
raw_data_path="/Volumes/ecommerce/source_data/raw_data/brands/*.csv"
df=spark.read.format("csv").option("header","true").schema(brand_schema).load(raw_data_path)
df= df.withColumn("source_filename",f.col("_metadata.file_path"))\
    .withColumn("created_at",f.current_timestamp())

#Write data to delta table
df.write.format("delta")\
    .option("mergeSchema","true")\
    .mode("overwrite")\
    .saveAsTable(f"{catalog_name}.bronze.brz_brands")
```

**Load Category Dimension Table**

```python
# define schema for the table
category_schema = StructType([
    StructField("category_code", StringType(), False),
    StructField("category_name", StringType(), True)])

#Read data from file
raw_data_path="/Volumes/ecommerce/source_data/raw_data/category/*.csv"
df=spark.read.format("csv").option("header","true").schema(category_schema).load(raw_data_path)
df= df.withColumn("source_filename",f.col("_metadata.file_path"))\
    .withColumn("created_at",f.current_timestamp())

#Write data to delta table
df.write.format("delta")\
    .option("mergeSchema","true")\
    .mode("overwrite")\
    .saveAsTable(f"{catalog_name}.bronze.brz_category")
```

**Load Customer Dimension Table**

```python
# Define schema for customer table
customer_schema = StructType([
    StructField("customer_id", StringType(), False),
    StructField("phone", StringType(), True),
    StructField("country_code", StringType(), True),
    StructField("country", StringType(), True),
    StructField("state", StringType(), True)
])
# Read data from file
df= spark.read.format("csv").option("header","true").schema(customer_schema).load("/Volumes/ecommerce/source_data/raw_data/customers/*.csv")
df= df.withColumn("source_filename",f.col("_metadata.file_path"))\
    .withColumn("created_at",f.current_timestamp())
#Write data to delta table
df.write.format("delta")\
    .option("mergeSchema","true")\
    .mode("overwrite")\
    .saveAsTable(f"{catalog_name}.bronze.brz_customer") 
```

**Load Date Dimension Table**

```python
# Define schema for Date table
date_schema= StructType([
  StructField("date", StringType(), True),
  StructField("year", IntegerType(), True),
  StructField("day_name", StringType(), True),
  StructField("quarter", IntegerType(), True),
  StructField("week_of_year", IntegerType(), True)
])
# Read from file
df= spark.read.format("csv").option("header","true").schema(date_schema).load("/Volumes/ecommerce/source_data/raw_data/date/*.csv")
df= df.withColumn("source_filename",f.col("_metadata.file_path"))\
  .withColumn("created_at",f.current_timestamp())

#Write data to delta table
df.write.format("delta")\
    .mode("overwrite")\
    .option("mergeSchema","true")\
    .saveAsTable(f"{catalog_name}.bronze.brz_date") 
```

**Load Product Dimension Table**

```python
# define schema for product table
product_schema = StructType([
    StructField("product_id", StringType(), False),
    StructField("sku", StringType(), True),
    StructField("category_code", StringType(), True),
    StructField("brand_code", StringType(), True),
    StructField("color", StringType(), True),
    StructField("size", StringType(), True),
    StructField("material", StringType(), True),
    StructField("weight_grams", StringType(), True),
    StructField("length_cm", StringType(), True),
    StructField("width_cm", FloatType(), True),
    StructField("height_cm", FloatType(), True),
    StructField("rating_count", IntegerType(), True)
])
# read data from file
df= spark.read.format("csv").option("header","true").schema(product_schema).load("/Volumes/ecommerce/source_data/raw_data/products/*.csv")
df= df.withColumn("source_filename",f.col("_metadata.file_path"))\
    .withColumn("created_at",f.current_timestamp())

#Write data to delta table
df.write.format("delta")\
    .option("MergeSchema","true")\
    .mode("overwrite")\
    .saveAsTable(f"{catalog_name}.bronze.brz_product") 
```
